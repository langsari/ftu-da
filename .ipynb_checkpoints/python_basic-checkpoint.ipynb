## การเขียนหมายเหตุ  
การเขียนหมายเหตุ (Comment) เป็นการเขียนข้อความลงในซอร์สโค๊ดของโปรแกรม นิยมเขียนเพื่ออธิบายความหมายภาย โดยที่หมายเหตุจะไม่ถูกนำไปประมวลผล
การเขียนหมายเหตุด้วยภาษาไพธอน ใช้เครื่องหมาย # (sharp) ไว้ด้านหน้าข้อความที่ต้องการ เช่น

```python
# นี่คือหมายเหตุ เขียนไว้ในโปรแกรม
a = 34,000      # ตัวแปร a ใช้เก็บเงินเดือน
print(a)        # แสดงค่าเงินเดือนในตัวแปร a ออกมา
```

## ตัวแปร (Variable)

ตัวแปร หมายถึง ชื่อที่กำหนดขึ้นมาเพื่อเก็บข้อมูลภายในโปรแกรม 
1. ประกอบด้วยตัวอักษร, ตัวอักษรปนตัวเลข หรือตัวอักษรปนเครื่องหมาย underscore (_) python > 3.0 สามารถประกาศตัวแปรด้วยภาษาต่าง ๆ ได้
2. ตัวแรกของชื่อตัวแปรต้องเป็นตัวอักษรห้ามขึ้นต้นด้วยตัวเลข
3. ตัวแปรมีความยาวอย่างน้อย 1 ตัวอักษร 
4. ห้ามใช้เครื่องหมายทางคณิตศาสตร์ในการตั้งชื่อตัวแปร
5. ห้ามตั้งชื่อซ้ำกับคำสงวน (Reserved word)
6. ตัวเลขและตัวใหญ่เป็นคนละตัวกัน (Case Sensitive)

### ตัวแปรชนิดเลขจำนวนเต็ม (integer)
ตัวแปรชนิดเลขจำนวนเต็มใช้เก็บค่าที่เป็นตัวเลขจำนวนเต็ม เช่น a = 3; b = 2; c=a+b

### ตัวแปรชนิดเลขทศนิยม (floating point)
ตัวแปรชนิดเลขทศนิยมใช้เก็บค่าที่เป็นตัวเลขทศนิยม เช่น pi = 3.1415

### ตัวแปรชนิดสตริงหรือตัวอักษร (String) 
ตัวแปรชนิดสตริงใช้เก็บข้อความ  สามารถเขียนได้ดังนี้  
1. การเปิดปิดข้อความด้วยเครื่องหมาย คำพูด (quote)  เช่น a = "Google Gmail Youtube"
2. การเปิดปิดข้อความด้วยเครื่องหมาย  single quote เช่น  a = 'Google Gmail Youtube'
3. การเปิดปิดด้วย triple quote เช่น a = """Google Gmail Youtube""" หรือ a = '''Google Gmail Youtube'''
จงตอบคำถามว่า ประกาศตัวแปรสตริงถูกต้องหรือไม่ ?

```python
a = "It's hot today."
b = 't's hot today.'
c = """t's hot today."""
d = '''t's hot today.'''
```

### ตัวแปรชนิดลีสต์ (List)
อาร์เรย์  
อาร์เรย์ คือ การประกาศตัวแปรด้วยชื่อเดียวกันและเข้าถึงสมาชิกด้วยหมายเลขดัชนี (index)  

การประกาศตัวแปรลีสต์ใช้เครื่องหมาย backet ([ ]) เปิดปิดมีรูปแบบคือ ชื่อตัวแปร  = [ ข้อมูล, ข้อมูล, ข้อมูล ] 

#### การเข้าถึงสมาชิกภายในตัวแปรลีสต์ (list)
1) เลขบวก เป็นการเข้าถึงสมาชิกจากซ้ายไปขวา  
2) เลขลบ เป็นการเข้าถึงสมาชิกจากขวามาซ้าย  
3) การเข้าถึงสมาชิกแบบช่วง ให้ระบุเลขลำดับเริ่มต้นและลำดับปลายทาง โดยคั่นด้วยเครื่องหมาย colon ( : ) เช่น a = [1,2,3,4,5]; a[1:3] จะได้ค่า [2,3] เป็นต้น  

```python
a = [5, 3, 6, 7, 2]
print(a)        # ผลลัพธ์ คือ ..............................
print(a[0])     # ผลลัพธ์ คือ ..............................
print(a[-1])    # ผลลัพธ์ คือ ..............................

b = [ 'Google', 'Gmail', 'Youtube' ]
print(b)        # ผลลัพธ์ คือ ..............................
print(b[0])     # ผลลัพธ์ คือ ..............................
print(b[-1])    # ผลลัพธ์ คือ ..............................
print(b[-2])    # ผลลัพธ์ คือ ..............................

c = [ "Google", 3, "Gmail", 8 ]
print(b)        # ผลลัพธ์ คือ ..............................
print(b[0])     # ผลลัพธ์ คือ ..............................
print(b[-1])    # ผลลัพธ์ คือ ..............................
print(b[-2])    # ผลลัพธ์ คือ ..............................

z = [ a, b, c ]
print(z[0])     # ผลลัพธ์ คือ ..............................
print(z[-1])    # ผลลัพธ์ คือ ..............................
print(z[1])     # ผลลัพธ์ คือ ..............................
print(z[-2])    # ผลลัพธ์ คือ ..............................

m = list(range(13))
print(m[0])     # ผลลัพธ์ คือ ..............................
print(m[4])     # ผลลัพธ์ คือ ..............................
print(m[1:2])   # ผลลัพธ์ คือ ..............................
print(m[2:4])   # ผลลัพธ์ คือ ..............................
```

#### การเรียงลำดับตัวแปรลีสต์ใช้คำสั่ง
sort()

```python
x = [ 30, 20, 15, 18, 14, 12 ]
print(x)    # ผลลัพธ์ คือ ..............................
x.sort()
print(x)    # ผลลัพธ์ คือ ..............................
```

#### การหาค่าสูงสุดและต่ำสุดในตัวแปรลีสต์
คำสั่ง max() และ min() ใช้สำหรับหาค่าที่สูงสูงสุดและต่ำสุดในตัวแปรลีสต์

```python
x = [ 30, 20, 15, 18, 14, 12 ]
print(max(x))   # ผลลัพธ์ คือ ..............................
print(min(x))   # ผลลัพธ์ คือ ..............................
```

#### การหาขนาดของตัวแปรลีสต์
 คำสั่ง len() ใช้สำหรับหาขนาดของตัวแปรลีสต์

#### การนับจำนวนข้อมูล
คำสั่ง count() ใช้สำหรับหาขนาดของตัวแปรลีสต์

```python
x = [ 5, 3, 4, 1, 3, 2, 4, 2, 4, 5, 4, 1, 1, 5 ]
print(x.count(3))   # ผลลัพธ์ คือ ..............................
print(x.count(5))   # ผลลัพธ์ คือ ..............................
print(x.count(1))   # ผลลัพธ์ คือ ..............................
print(len(x))       # ผลลัพธ์ คือ ..............................
```

#### การรวมข้อมูล
คำสั่ง sum() ใช้สำหรับคำนวณผลรวมภายในตัวแปรลีสต์ ใช้สำหรับกรณีที่ตัวแปรในลีสต์เป็นค่าตัวเลขเท่านั้น

```python
x = [ 5, 3, 4, 1, 3, 2, 4, 2, 4, 5, 4, 1, 1, 5 ]
print(sum(x))   # ผลลัพธ์ คือ ..............................
```
